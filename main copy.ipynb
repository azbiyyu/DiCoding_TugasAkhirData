========= APA KATEGORI PRODUK TERPOPULER ? ==========
Pertanyaan: "Apa saja kategori produk yang paling populer dalam hal jumlah produk yang terjual?"
Analisis yang Diperlukan: 
1. menghitung total produk yang terjual dalam setiap kategori 2. mengidentifikasi kategori produk yang memiliki penjualan tertinggi.
Langkah-langkah Analisis:
1. Kelompokkan data berdasarkan 'product_category_name'.
2. Hitung jumlah produk yang terjual dalam setiap kategori.
3. Identifikasi kategori dengan jumlah produk terjual tertinggi.

```python
# MENGELOMPOKKAN DATA
kategori_penjualan = df.groupby('product_category_name')['product_id'].count().reset_index()
kategori_penjualan.columns = ['product_category_name','jumlah_produk_terjual']
```

```python
kategori_terlaris = kategori_penjualan[kategori_penjualan['jumlah_produk_terjual'] == kategori_penjualan['jumlah_produk_terjual'].max()]
```

```python
# MEMBUAT VISUALISAI
plt.figure(figsize=(12,6))
plt.bar(kategori_penjualan['product_category_name'],kategori_penjualan['jumlah_produk_terjual'])
plt.title('Jumlah Produk Terjual per Kategori')
plt.xlabel('Kategori Produk')
plt.xticks(rotation= 90, ha = 'right')
plt.ylabel('Jumlah Produk Terjual')
plt.show()
```

```python
# menampilkan hasil
print("1. Kelompokkan data berdasarkan 'product_category_name' dan hitung jumlah produk terjual dalam setiap kategori:")
print(kategori_penjualan)
```

```python
print("\n2. Identifikasi kategori dengan jumlah produk terjual tertinggi:")
print(kategori_terlaris)
```

==== APA HUBUNGAN ANTARA DIMENSI PROUK DAN JUMLAH FOTO ? ====
Pertanyaan: "Apakah ada hubungan antara dimensi fisik produk (panjang, tinggi, lebar) dengan jumlah foto yang diunggah?"
Analisis yang Diperlukan: 
1. menguji apakah dimensi fisik produk memiliki korelasi dengan jumlah foto yang diunggah.
Langkah-langkah Analisis:
2. Hitung koefisien korelasi antara dimensi produk dan jumlah 3. foto yang diunggah.
4. Buat scatter plot untuk memvisualisasikan hubungan antara dimensi produk dan jumlah foto.

```python
# hitung koefisien korelasi dimensi produk (height,width,length) dan jumlah foto
dimension = df[['product_length_cm','product_height_cm','product_width_cm']]
product_photos_qty = df['product_photos_qty']
print(df['product_photos_qty'].unique())
```

```python
correlation_matrix = dimension.corrwith(product_photos_qty)
```

```python
# Scatter plot visualisasi korelasi dimensi produk dan jumlah foto
plt.figure(figsize=(12,6))
# PANJANG dan JUMLAH FOTO
plt.scatter(df['product_length_cm'], product_photos_qty, alpha=0.5)
plt.title('Panjang Produk vs Jumlah Foto')
plt.xlabel('Panjang Produk (cm)')
plt.ylabel('Jumlah Foto')
plt.show()
```

```python
# TINGGI dan JUMLAH FOTO
plt.scatter(df['product_height_cm'], product_photos_qty, alpha=0.5)
plt.title('Tinggi Produk vs Jumlah Foto')
plt.xlabel('Tinggi Produk (cm)')
plt.ylabel('Jumlah Foto')
plt.show()
```

```python
# LEBAR dan JUMLAH FOTO
plt.scatter(df['product_width_cm'], product_photos_qty, alpha=0.5)
plt.title('Lebar Produk vs Jumlah Foto')
plt.xlabel('Lebar Produk (cm)')
plt.ylabel('Jumlah Foto')
plt.show()
```

```python
# Tampilkan koefisien korelasi
print("Koefisien Korelasi antara Dimensi Produk dan Jumlah Foto:")
print(correlation_matrix)
```

```python
# COBA MEMBUAT YANG VOLUME (korelasi volume dan jumlah foto)
# Menghitung volume produk
df['product_volume_cm3'] = (
    df['product_length_cm'] *
    df['product_height_cm'] *
    df['product_width_cm']
)
```

```python
# VOLUME dan JUMLAH FOTO
plt.scatter(df['product_volume_cm3'], product_photos_qty, alpha=0.5)
plt.title('Volume Produk vs Jumlah Foto')
plt.xlabel('Volume Produk (cm3)')
plt.ylabel('Jumlah Foto')
plt.show()
```

```python
# Menghitung korelasi antara volume produk dan jumlah foto yang diunggah
correlation = df['product_photos_qty'].corr(df['product_volume_cm3'])

print("Korelasi antara Dimensi Produk (Volume) dan Jumlah Foto:", correlation)
```

```python
# Gabungkan dengan data 'product_photos_qty'
df_merged = df.merge(kategori_penjualan, on='product_category_name', how='left')
df_merged.head()
```

```python
plt.figure(figsize=(12, 6))

# Histogram untuk 'product_photos_qty'
plt.hist(df_merged['product_photos_qty'], bins=20, alpha=0.5, color='blue', label='Jumlah Foto')
plt.axvline(df_merged['product_photos_qty'].mean(), color='blue', linestyle='dashed', linewidth=2, label='Rata-rata Jumlah Foto')

# Histogram untuk 'jumlah_produk_terjual'
plt.hist(df_merged['jumlah_produk_terjual'], bins=20, alpha=0.5, color='green', label='Jumlah Produk Terjual')
plt.axvline(df_merged['jumlah_produk_terjual'].mean(), color='green', linestyle='dashed', linewidth=2, label='Rata-rata Jumlah Produk Terjual')

plt.title('Histogram Hubungan Jumlah Foto dan Jumlah Produk Terjual')
plt.xlabel('Jumlah')
plt.ylabel('Frekuensi')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menambahkan anotasi rata-rata ke grafik
plt.annotate(f"Rata-rata Jumlah Foto: {df_merged['product_photos_qty'].mean():.2f}",
             xy=(df_merged['product_photos_qty'].mean(), 0),
             xytext=(df_merged['product_photos_qty'].mean(), 3000),
             arrowprops=dict(facecolor='blue', arrowstyle='->'))

plt.annotate(f"Rata-rata Jumlah Produk Terjual: {df_merged['jumlah_produk_terjual'].mean():.2f}",
             xy=(df_merged['jumlah_produk_terjual'].mean(), 0),
             xytext=(df_merged['jumlah_produk_terjual'].mean(), 3000),
             arrowprops=dict(facecolor='green', arrowstyle='->'))

plt.show()
```

```python
# Kelompokkan data berdasarkan 'product_photos_qty' dan hitung jumlah produk terjual dalam setiap kelompok jumlah foto
foto_produk_terjual = df.groupby('product_photos_qty')['product_id'].count().reset_index()
foto_produk_terjual.columns = ['product_photos_qty', 'jumlah_produk_terjual']

# Buat histogram
plt.figure(figsize=(12, 6))

# Buat larik angka dari 0 hingga 50
jumlah_foto = np.arange(51)

plt.bar(foto_produk_terjual['product_photos_qty'], foto_produk_terjual['jumlah_produk_terjual'], color='skyblue')
plt.xticks(jumlah_foto)  # Atur label sumbu X dengan angka dari 0 hingga 50
plt.xlim(-1, 51)  # Atur batas sumbu X dari -1 hingga 51
plt.xlabel('Jumlah Foto')
plt.ylabel('Jumlah Produk Terjual')
plt.title('Hubungan antara Jumlah Foto dan Jumlah Produk Terjual')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
